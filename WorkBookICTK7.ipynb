# Artificial Neural Networks.

*  Bob started a mobile company and wants to compete with big brands.
*  He collected mobile sales data to understand how features affect price.
*  Build an ANN classifier using mobile_price_classification.csv.
+  Label: price_range
*  Features: All other columns
*  Perform hyperparameter tuning.



## Load and explore the data


```python
import pandas as pd

df = pd.read_csv("/content/sample_data/mobile_price_classification.csv")

display(df.head())
display(df.info())
display(df.describe())
```

## Preprocess the data



```python
from sklearn.preprocessing import StandardScaler

X = df.drop('price_range', axis=1)
y = df['price_range']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

display(X_scaled[:5])
```

## Build the neural network model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

# Add the input layer and the first hidden layer
model.add(Dense(units=64, activation='relu', input_shape=(X_scaled.shape[1],)))

# Add one or more hidden layers
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))

# Add the output layer
model.add(Dense(units=4, activation='softmax'))

# Display the model summary
model.summary()
```

## Train the model

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_scaled, y, epochs=50, batch_size=32)
```

## Evaluate the model


```python
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

y_pred_prob = model.predict(X_scaled)
y_pred = np.argmax(y_pred_prob, axis=1)

accuracy = accuracy_score(y, y_pred)
print(f"Overall Accuracy: {accuracy:.4f}")

report = classification_report(y, y_pred)
print("Classification Report:")
print(report)
```

## Hyperparameter tuning


```python
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split

def build_model(hp):
    model = Sequential()
    model.add(Dense(units=hp.Int('units_input', min_value=32, max_value=128, step=32),
                    activation='relu',
                    input_shape=(X_scaled.shape[1],)))
    model.add(Dropout(rate=hp.Float('dropout_input', min_value=0.0, max_value=0.5, step=0.1)))

    # Add one or more hidden layers
    for i in range(hp.Int('num_hidden_layers', min_value=1, max_value=3)):
        model.add(Dense(units=hp.Int(f'units_hidden_{i}', min_value=16, max_value=64, step=16),
                        activation='relu'))
        model.add(Dropout(rate=hp.Float(f'dropout_hidden_{i}', min_value=0.0, max_value=0.5, step=0.1)))

    # Add the output layer
    model.add(Dense(units=4, activation='softmax'))

    # Compile the model
    model.compile(optimizer=hp.Choice('optimizer', values=['adam', 'sgd', 'rmsprop']),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    return model
```

```python
# Split data for validation during tuning
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Instantiate the tuner
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,
    executions_per_trial=2,
    directory='my_dir',
    project_name='keras_tuning'
)

# Start the hyperparameter search
tuner.search(X_train, y_train, epochs=10, validation_data=(X_val, y_val))
```

**Retrieve the best hyperparameters found by the tuner and display them.**


```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

print(f"Best Hyperparameters:")
print(f"Input layer units: {best_hps.get('units_input')}")
print(f"Input layer dropout: {best_hps.get('dropout_input')}")
print(f"Number of hidden layers: {best_hps.get('num_hidden_layers')}")
for i in range(best_hps.get('num_hidden_layers')):
    print(f"Hidden layer {i} units: {best_hps.get(f'units_hidden_{i}')}")
    print(f"Hidden layer {i} dropout: {best_hps.get(f'dropout_hidden_{i}')}")
print(f"Optimizer: {best_hps.get('optimizer')}")
```

## Retrain with best parameters



```python
best_model = tuner.get_best_models(num_models=1)[0]

best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
best_optimizer = best_hps.get('optimizer')

best_model.compile(optimizer=best_optimizer,
                   loss='sparse_categorical_crossentropy',
                   metrics=['accuracy'])

history = best_model.fit(X_scaled, y, epochs=50, batch_size=32)
```

## Final evaluation


```python
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

y_pred_prob = best_model.predict(X_scaled)
y_pred = np.argmax(y_pred_prob, axis=1)

accuracy = accuracy_score(y, y_pred)
print(f"Overall Accuracy: {accuracy:.4f}")

report = classification_report(y, y_pred)
print("Classification Report:")
print(report)
```
