# NumPy (Array Operations)

1. Create a NumPy array of shape (5, 5) with random integers between 1 and 50.
* Find the row with the maximum sum.
* Replace all elements greater than the mean with 0.

```python
import numpy as np #importing numpy library
arr=np.random.randint(1,50,(5,5)) #creating a 5*5 array with random integer 1 to 50
print("Array: ")
print(arr)

# Find the row with the maximum sum
rsum=np.sum(arr,axis=1) #Find row wise sum
print("Row sum: ",rsum)
print("Raw with maximum sum: ",np.argmax(rsum)) #Raw number with maximum sum

# Replace all elements greater than the mean with 0
print("Mean: ",np.mean(arr)) #Mean of array
arr1=np.where(arr>np.mean(arr),0,arr) #Replace all elements greater than the mean with 0
print("New array: ")
print(arr1)
```

2. Create a NumPy array with numbers from 1 to 20. Select every 3rd element starting from the 2nd element and store them in a new array.

```python
arr=np.arange(1,21) #creating a array with numbers from 1 to 20
print("Array: ",arr)
arr1=arr[1::3] #Slicing every 3 rd element starting from 2nd
print("Selected array: ",arr1)
```

# Pandas (Data Handling)

3. Load the dataset Iris data
* Display the first 10 rows and basic statistics.
* Filter all rows where **sepal_length > median and species is "versicolor"**.
* Group the data by species and compute the mean petal_length.

```python
import pandas as pd
import seaborn as sns

# Load the Iris dataset
df=sns.load_dataset('iris')
print("First 10 rows are: ")
print(df.head(10)) #first 10 rows
print("Basic statistics: ")
print(df.describe()) #basic statistics

#filter all rows where sepal_length > median and species is "versicolor"
fltr=df[( (df['sepal_length'] > df['sepal_length'].median()) & (df['species'] == "versicolor") )]
print("Filtered data: ")
print(fltr)

#Group the data by species and compute the mean petal_length
grp=df.groupby('species')['petal_length'].mean()
print("Grouped data: ")
print(grp)
```

4. Add a new column named petal_ratio = petal_length / petal_width.
* Find the top 5 rows with the highest petal_ratio.



```python
# Add a new column named petal_ratio = petal_length / petal_width.
df['petal_ratio'] = df['petal_length'] / df['petal_width']
top_5_petal_ratio = df.nlargest(5, 'petal_ratio') # Find the top 5 rows with the highest petal_ratio.
print("Top 5 rows with the highest petal_ratio:")
display(top_5_petal_ratio)
```

# Data Visualization

5. Using Matplotlib, create:
* A histogram of sepal_length with 15 bins.
* A scatter plot of sepal_length vs petal_length, color-coded by species.

```python
import matplotlib.pyplot as plt

#A histogram of sepal_length with 15 bins.
plt.hist(df['sepal_length'],bins=15)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()

#A scatter plot of sepal_length vs petal_length, color-coded by species.
colors = df['species'].map({'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'})
plt.scatter(df['sepal_length'], df['petal_length'], c=colors)
plt.title('Scatter Plot of Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

6. Using Seaborn, plot:
* A pairplot of the entire dataset.
* A boxplot showing petal_width distribution for each species.




```python
import seaborn as sns

#pairplot of entier dataset
sns.pairplot(df)
plt.show()

#boxplot showing petal_width distribution for each species
sns.boxplot(x='species',y='petal_width',data=df)
plt.show()
```
